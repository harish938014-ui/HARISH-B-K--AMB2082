```python
for i in range(3):
  for j in range(3):
    print("*",end="")
  print()
```

```python
rows=int(input("enter rows:"))
cols=int(input("enter cols:"))
for i in range (rows):
  for j in range (cols):
    print("#",end="")
  print()
```

```python
for i in range (3):
  j =0
while j<2:
  print(i,j)
  j+=1
```

```python
for i in range (2):
  for j in range(3):
    print("i=",i,"j=",j)
    print(i,j)
```
