```python
import pandas as pd
# Farm dataset with many tracked factors
data = {
 "Rainfall_mm": [200, 180, 220, 210, 190, 230],
 "Soil_Moisture": [30, 28, 35, 33, 29, 36],
 "Humidity": [80, 75, 85, 82, 78, 88],
 "Wind_Speed": [10, 12, 9, 11, 10, 8],
 "Yield_kg": [3200, 3000, 3500, 3400, 3100, 3600]
}
df = pd.DataFrame(data)
df
```

```python
from sklearn.ensemble import RandomForestRegressor
X = df.drop("Yield_kg", axis=1)
y = df["Yield_kg"]
model = RandomForestRegressor(random_state=42)
model.fit(X, y)
```

```python
importances = model.feature_importances_
feature_importance = pd.Series(
 importances,
 index=X.columns
)
feature_importance
```

```python
importances = model.feature_importances_
feature_importance = pd.Series(
 importances,
 index=X.columns
)
feature_importance.sort_values(ascending=False)
```

```python
importances = model.feature_importances_
feature_importance = pd.Series(
 importances,
 index=X.columns
)
feature_importance.sort_values(ascending=False)
feature_importance.plot(
 kind="barh",
 title="Feature Importance for Yield Prediction"
 )
```
