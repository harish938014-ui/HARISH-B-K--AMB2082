```python
daily_rainfall = np.array([0.2, 0.0, 0.5, 1.2, 0.3, 0.1, 0.0])
# Sum
total = np.sum(daily_rainfall)
# Output: 2.3
# Mean (average)
average = np.mean(daily_rainfall)
# Output: 0.328...
# Standard deviation
variability = np.std(daily_rainfall)
# Output: 0.404...
# Min and Max
driest = np.min(daily_rainfall)  # 0.0
wettest = np.max(daily_rainfall)  # 1.2
print(total)
print(average)
print(variability)
print(driest)
print(wettest)
```

```python
# Crop yields: 3 fields × 4 years
yields = np.array([
    [120, 125, 130, 135],
    [115, 120, 122, 128],
    [130, 135, 140, 142]
])
# Average across all data
overall_avg = np.mean(yields)

# Average per field (across years)
field_avgs = np.mean(yields, axis=1)

# Average per year (across fields)
year_avgs = np.mean(yields, axis=0)
print(overall_avg)
print(field_avgs)
print(year_avgs)
```

```python
# Simulated hourly temperature data (4 sensors × 24 hours)
np.random.seed(42) # For reproducibility
sensors = np.random.normal(loc=75, scale=5, size=(4, 24))
# Round to 1 decimal place
sensors = np.round(sensors, 1)
# Problem 1: Find all readings above 82°F (danger threshold)
overheating = sensors > 82
critical_temps = sensors[overheating]
print(f"Critical readings: {len(critical_temps)} occurrences")
print(f"Hottest reading: {np.max(sensors):.1f}°F")
# Problem 2: Calculate statistics per sensor
for i in range(4):
 sensor_data = sensors[i, :]
 print(f"Sensor {i+1}:")
 print(f" Average: {np.mean(sensor_data):.1f}°F")
 print(f" Range: {np.min(sensor_data):.1f}°F to {np.max(sensor_data):.1f}°F")
 print(f" Std Dev: {np.std(sensor_data):.1f}°F")
# Problem 3: Calculate hourly averages (across all sensors)
hourly_avg = np.mean(sensors, axis=0)
hottest_hour = np.argmax(hourly_avg)
print(f"Hottest hour: Hour {hottest_hour} ({hourly_avg[hottest_hour]:.1f}°F")
```
