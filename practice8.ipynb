```python
import numpy as np
#Soil pH readings (3 fields × 4 samples)
ph_levels = np.array([
    [6.5, 6.8, 6.3, 6.7],
    [7.1, 7.0, 6.9, 7.2],
    [6.2, 6.4, 6.1, 6.3]
])
# Single element: row 1, column 2
print(ph_levels[1, 2])
# Entire row 0
print(ph_levels[0, :])

# Entire column 3
print(ph_levels[:, 3])
```

```python
import numpy as np
temperatures = np.array([72, 85, 91, 68, 95, 70, 88])
# Create a boolean array (mask)
hot_days = temperatures > 85
print(hot_days)
# Output: [False False True False True False True]
# Use mask to filter original array
print(temperatures[hot_days])

# Can do it in one line
print(temperatures[temperatures > 85])
```

```python
import numpy as np
# Crop yields in tons
field_a = np.array([120, 135, 142, 128])
field_b = np.array([118, 140, 138, 132])
# Add arrays
total_yield = field_a + field_b
print(total_yield)
# Subtract to find difference
difference = field_a - field_b
print(difference)
# Multiply each by a constant
doubled = field_a * 2
print(doubled)
```

```python
import numpy as np
temperatures = np.array([72, 85, 91, 68, 95, 70, 88])
# Combine conditions with & (and), | (or)
moderate = temperatures[(temperatures >= 70) &
(temperatures <= 85)]
print(moderate)
# Find extreme values (very hot OR very cold)
extreme = temperatures[(temperatures > 90) |
(temperatures < 70)]
print(extreme)
```

```python
import numpy as np
daily_rainfall = np.array([0.2, 0.0, 0.5, 1.2, 0.3, 0.1, 0.0])
# Sum
total = np.sum(daily_rainfall)

# Mean (average)
average = np.mean(daily_rainfall)
# Standard deviation
variability = np.std(daily_rainfall)
# Min and Max
driest = np.min(daily_rainfall)
wettest = np.max(daily_rainfall)
print(total)
print(average)
print(variability)
print(driest)
print(wettest)
```

```python
import numpy as np
# Crop yields: 3 fields × 4 years
yields = np.array([
    [120, 125, 130, 135],
    [115, 120, 122, 128],
    [130, 135, 140, 142]
])
# Average across all data
overall_avg = np.mean(yields)
# Average per field (across years)
field_avgs = np.mean(yields, axis=1)
# Average per year (across fields)
year_avgs = np.mean(yields, axis=0)
print(overall_avg)
print(field_avgs)
print(year_avgs)
```

```python
# Simulated hourly temperature data (4 sensors × 24 hours)
np.random.seed(42) # For reproducibility
sensors = np.random.normal(loc=75, scale=5, size=(4, 24))
# Round to 1 decimal place
sensors = np.round(sensors, 1)
# Problem 1: Find all readings above 82°F (danger threshold)
overheating = sensors > 82
critical_temps = sensors[overheating]
print(f"Critical readings: {len(critical_temps)} occurrences")
print(f"Hottest reading: {np.max(sensors):.1f}°F")
# Problem 2: Calculate statistics per sensor
for i in range(4):
 sensor_data = sensors[i, :]
 print(f"Sensor {i+1}:")
 print(f" Average: {np.mean(sensor_data):.1f}°F")
 print(f" Range: {np.min(sensor_data):.1f}°F to {np.max(sensor_data):.1f}°F")
 print(f" Std Dev: {np.std(sensor_data):.1f}°F")
# Problem 3: Calculate hourly averages (across all sensors)
hourly_avg = np.mean(sensors, axis=0)
hottest_hour = np.argmax(hourly_avg)
print(f"Hottest hour: Hour {hottest_hour} ({hourly_avg[hottest_hour]:.1f}°F)")
```
